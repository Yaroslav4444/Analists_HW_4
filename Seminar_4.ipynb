## 1.1. Скачать данные в закрепе к уроку

```python
df = pd.read_csv('kc_house_data.csv')
df
```

## 1.2 Изучите стоимости недвижимости

```python
plt.figure(figsize=(8, 6))
plt.plot(df['price'], df['sqft_living'])
plt.xlabel('Цена')
plt.ylabel('Площадь')
plt.title('Стоимость недвижимости')
plt.grid();
```

Вывод: цена зависит от площади.

## 1.3 Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['sqft_living'], bins=10)
plt.xlabel('Жилая площадь, кв')
plt.ylabel('Цена')
plt.title('Жилая площадь');
```

## 1.4 Изучите распределение года постройки

```python
data = df['yr_built'].value_counts()
data
```

```python
plt.figure(figsize=(8, 6))
plt.pie(data.values, autopct='%1.1f%%', labels=data.index)
plt.title('Годы постройки');
```

Вывод: распределение равномерное.

# 3 задача
## Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
## Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg')
plt.title('Влияние жилплощади на цену');
```

Вывод: на цену жилплощади сильно влияет размер помещения.

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg')
plt.title('Влияние категории недвижимости на цену');
```

Вывод: от категории недвижимости сильно меняется цена.

```python
sns.jointplot(x=df['price'], y=df['bedrooms'], kind='reg')
plt.title('Влияние кол-ва спален на цену недвижимости');
```

Вывод: чем больше спален,тем дороже квартира.

```python
sns.jointplot(x=df['price'], y=df['bathrooms'], kind='reg')
plt.title('Влияние кол-ва ванных комнат на цену недвижимости');
```

Вывод: кол-во ванных комнат может быть большим и в не дорогих квартирах.

```python
sns.jointplot(x=df['price'], y=df['yr_built'], kind='reg')
plt.title('Влияние года постройки на цену недвижимости');
```

Вывод: хорошо видна линейная зависимость цены от года постройки.
